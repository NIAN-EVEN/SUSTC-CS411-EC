```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
point_num = 1000
gaussian_1_dim_x = [np.random.standard_normal() for x in range(point_num)]
gaussian_1_dim_y = [np.random.standard_normal() for x in range(point_num)]
cauchy_1_dim_x = [np.random.standard_cauchy() for x in range(point_num)]
cauchy_1_dim_y = [np.random.standard_cauchy() for x in range(point_num)]
```

```python
plt.figure()
plt.scatter(gaussian_1_dim_x, gaussian_1_dim_y, marker=".", c="r")
plt.scatter(cauchy_1_dim_x, cauchy_1_dim_y, marker=".", c="g")
plt.show()
```
